# Module 0 — Environment Setup
This notebook sets up the AI Engineering development environment.
We install libraries, store API keys safely, and make our first API call to an LLM.

```python
import os
from google.colab import userdata

groq_api_key = userdata.get('GROQ_API_KEY').strip()
os.environ["GROQ_API_KEY"] = groq_api_key
print("API key loaded successfully!")
```

```python
# groq is the official Groq library we installed earlier
# Groq class lets us create a connection to the Groq API
from groq import Groq

# Create a client — think of this as opening a WhatsApp connection to the AI
# Pass the API key directly to ensure it's stripped correctly
client = Groq(api_key=groq_api_key)

# Send a message and get a reply back
# We are asking the AI a simple question about Swiggy
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",  # the AI model we want to use
    messages=[
        {"role": "user", "content": "What is Swiggy and how does it work?"}
    ]
)

print(response)
```

```python
# Extract just the actual reply text from the response
# choices[0] means the first reply (AI can give multiple choices, we want the first)
# message.content is the actual text the AI wrote
answer = response.choices[0].message.content
print("AI Reply:")
print(answer)
```

```python
# Extract token usage information
# Tokens are like "word pieces" — how the AI measures text length
# prompt_tokens = tokens in our question
# completion_tokens = tokens in the AI's reply
# total_tokens = both combined — this is what determines your API cost

print("Model used:", response.model)
print("Tokens in our question:", response.usage.prompt_tokens)
print("Tokens in AI reply:", response.usage.completion_tokens)
print("Total tokens used:", response.usage.total_tokens)
```

```python
# Clean summary of our first API call
print("=" * 40)
print("MODULE 0 — FIRST API CALL SUMMARY")
print("=" * 40)
print(f"Model      : {response.model}")
print(f"Question   : What is Swiggy and how does it work?")
print(f"Prompt tokens    : {response.usage.prompt_tokens}")
print(f"Reply tokens     : {response.usage.completion_tokens}")
print(f"Total tokens     : {response.usage.total_tokens}")
print(f"Time taken       : {response.usage.total_time:.2f} seconds")
print("=" * 40)
```
